```python
# Junior and Senior lessions
folder_name_s = "Senior"
folder_name_j = "Junior"
# folder_name_e = "Elementary"
senior_index = 1
junior_index = 1

for i in range(range_j_s):
    base_link_senior = f"https://apostolicfaithweca.org/sunday-school-lesson-library?llang=1&slc=1&title=&Slno={senior_index}"
    base_link_junior = f"https://apostolicfaithweca.org/sunday-school-lesson-library?llang=1&slc=2&title=&Slno={junior_index}"
    print(base_link_senior)
    print(base_link_junior)
    print()
    r_s = requests.get(base_link_senior)
    r_j = requests.get(base_link_junior)
    soup_s = bs(r_s.content)
    soup_j = bs(r_j.content)
    senior_suffix = soup_s.find_all(class_ = "views-field views-field-title")[1].a['href']
    junior_suffix = soup_j.find_all(class_ = "views-field views-field-title")[1].a['href']
    
    main_senior = requests.get(base_link+senior_suffix)
    main_junior = requests.get(base_link+junior_suffix)
    
    soup_main_s = bs(main_senior.content)
    soup_main_j = bs(main_junior.content)
    
    data_page_s = soup_main_s.find_all(class_ = "col-sm-12 bs-region bs-region--main")[0]
    file_name_s = "".join(data_page_s.find_all("p")[0].text.split())
    file_name_s = "_".join(file_name_s.split("-"))
    
    with open("{}/{}.txt".format(folder_name_s, file_name_s), "w") as file1:
        file1.write("Bible Text: {}".format(data_page_s.find_all("div")[0].text.strip()))
        file1.write("\n\n")
        file1.write(data_page_s.find_all("p")[0].text)
        file1.write("\n\n")
        file1.write("{}: {}".format(data_page_s.find_all("h6")[0].text, data_page_s.find_all("div")[-1].text.strip()))
        file1.write("\n\n")
        file1.write(data_page_s.find_all("h6")[1].text.strip())
        # file.write("\n")
        for values in data_page_s.find_all("p")[1:]:
            file1.write(values.text.strip())
            file1.write("\n\n")
        file1.close()
    
    data_page_j = soup_main_j.find_all(class_ = "col-sm-12 bs-region bs-region--main")[0]
    file_name_j = "".join(data_page_j.find_all("p")[0].text.split())
    file_name_j = "_".join(file_name_j.split("-"))
    
    with open("{}/{}.txt".format(folder_name_j, file_name_j), "w") as file2:
        file2.write("Bible Text: {}".format(data_page_j.find_all("div")[0].text.strip()))
        file2.write("\n\n")
        file2.write(data_page_j.find_all("p")[0].text)
        file2.write("\n\n")
        file2.write("{}: {}".format(data_page_j.find_all("h6")[0].text, data_page_j.find_all("div")[-1].text.strip()))
        file2.write("\n\n")
        file2.write(data_page_j.find_all("h6")[1].text.strip())
        # file.write("\n")
        for values in data_page_j.find_all("p")[1:]:
            file2.write(values.text.strip())
            file2.write("\n\n")
        file2.close() 
    
    
    senior_index+=1
    junior_index+=1
```
